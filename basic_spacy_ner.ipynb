```python
import spacy
from spacy import displacy
```

```python
# models need to be installed first
# see https://spacy.io/models
nome_modello = "en_core_web_sm"

# https://www.balcanicaucaso.org/aree/Bulgaria/Elezioni-in-Bulgaria-per-Borisov-una-vittoria-complicata-232015
# distribuito sotto Creative Commons Attribuzione - Non commerciale - Non opere derivate 4.0 Internazionale
nome_file = "data/bulgaria_complicata.txt"
```

```python
with open(nome_file) as f:
    text = f.read()
```

```python
nlp = spacy.load(nome_modello)
doc = nlp(text)
```

```python
print(f'{"text":40}' f'{"label":22}' f'{"range":10}' f'{"explanation"}')
for ent in doc.ents:
    print(
        f"{str(ent.text):40}"
        f"{str(ent.label_):22}"
        f"{ent.start_char}-{ent.end_char}\t"
        f"{spacy.explain(ent.label_)}"
    )
```

```python
displacy.serve(doc, style="ent")
```
