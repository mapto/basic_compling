```python
from collections import Counter

import spacy
```

```python
# models need to be installed first
# see https://spacy.io/models
nome_modello = "en_core_web_sm"
nome_file = "data/Head_1666.txt"

n = 2
```

```python
with open(nome_file) as f:
    text = f.read()
```

```python
nlp = spacy.load(nome_modello)
doc = nlp(text)
```

```python
num_gram = Counter()
cgram = []
for token in doc:
    if token.is_punct:
        cgram = []
        continue
    if not token.is_alpha or token.is_stop:
        continue

    if len(cgram) == n - 1:
        num_gram[" ".join(cgram + [token.lower_])] += 1
    cgram = cgram[-(n - 1) :] + [token.lower_]

num_gram
```

```python
with open(f"{nome_file}.{n}grams.csv", "w") as fout:
    fout.write("Item, Frequency\n")
    fout.writelines([f'"{k}", {v}\n' for k, v in num_gram.most_common()])
```
